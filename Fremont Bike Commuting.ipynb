```python
URL = 'https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD'
```

```python
from urllib.request import urlretrieve
urlretrieve(URL,'Fremont.csv')
```

```python
import pandas as pd
#read file, change index column to date column, turn dates into proper format
data = pd.read_csv('Fremont.csv',index_col='Date',parse_dates=True)
data.head()
```

```python
#initial visual
%matplotlib inline
data.plot();

#sum at week level to declutter
data.resample('W').sum().plot()
```

```python
#change to beter looking style
import matplotlib.pyplot as plt
plt.style.use('seaborn')
data.columns = ('West','East')
data.plot();
#sum at week level
data.resample('W').sum().plot()
```

```python
#add a new Total Column
data['Total'] = data['West'] + data['East']

#annual trend - rolling year (so a datapoint is showing a sum of the last year)
ax = data.resample('D').sum().rolling(365).sum().plot();
ax.set_ylim(0,None);
```

```python
#look at time of day
data.groupby(data.index.time).mean().plot();

#notice pattern with more traffic on the East side in the morning,
#followed by the West side in the afternoon (WORK COMMUTE)
```

```python
pivoted = data.pivot_table('Total',index=data.index.time,columns=data.index.date)
pivoted.iloc[:5,:5]
```

```python
#line for each day of the year
#alpha adds transparency
pivoted.plot(legend=False,alpha=0.01)
```
